```python
# compute the k means

import random as rnd
def kmean(k,X,Y,ite):                                          # Function kmn(k) constructs k clusters centered initially at random elements of the set (X,Y)
                                                             # Nit is the number of iterations of the procedure
    mean_x  = sum(X) / len(X)
    sqrt_x  = np.sqrt(abs(sum(X **2 ) / len(X) - mean_x ** 2))
    mean_y  = sum(Y) / len(Y)
    sqrt_y  = np.sqrt(abs(sum(Y **2 ) / len(Y) - mean_y ** 2))
    Mx_proto  = np.zeros(k)                                        # Define the array for the x-positions of the first k-prototypes
    My_proto  = np.zeros(k)                                        # Define the array for the y-positions of the first k-prototypes

    for i in range(0,k):                                     # Fill in the Mx and My arrays with the randomly chosen data points
        Mx_proto[i] = mean_x + rnd.uniform(-sqrt_x,sqrt_x) / 1
        My_proto[i] = mean_y + rnd.uniform(-sqrt_y,sqrt_y) / 1

    szx = np.size(X)                                         # Measures the size of the data set

    for it in range(0,ite):                                  # Main iteration of the k-means algorithm

        grades = []                                         # Construct and empty list

        for i in range(0,k):
            grades.append([])                               # Stores k empty lists inside the first list. Inside each list will be plased the index of a
                                                             # datum correspondent to that class
        for n in range(0,szx):                               # Loop in data points
            width = np.zeros(k)                           # Vector where the distance of the data points to the prototypes is stored
            for i in range(0,k):                             # Loop in classes
                width[i] = (Mx_proto[i] - X[n]) ** 2 + (My_proto[i] - Y[n]) ** 2  # Computation of the Euclidian distance
            dd = width[0]
            kk = 0
            for i in range(0,k):
                if dd > width[i]:
                    kk = i
                    dd = width[i]

            grades[kk].append(n)                            # the closest, and then the index of the datum is stored in the correspondent class

        Mx_proto = np.zeros(k)                                     # Recomputation of the prototypes as the arithmetic means inside each class
        My_proto = np.zeros(k)
        for i in range(0,k):
            for n in grades[i]:
                Mx_proto[i] = Mx_proto[i] + X[n]
                My_proto[i] = My_proto[i] + Y[n]
            Mx_proto[i] = Mx_proto[i] / max(1,np.size(grades[i]))
            My_proto[i] = My_proto[i] / max(1,np.size(grades[i]))




        return(Mx_proto,My_proto,grades)
```

```python

def diff(cen_x,cen_y,cn_d):                               # Function TE(cx,cy,ind) computs the distance between prototypes and points in the class
    err = 0                                        # Error
    Lenth = len(cn_d)
    k = 0
    while k < Lenth:
        for i in cn_d[k]:                                          # Loop in datapoints
            err = err + (a[i] - cen_x[k]) ** 2 + (b[i] - cen_y[k]) ** 2   # Increment in Error
        k = k + 1
    return(err)                                                     # Returns the training error

klus_x = [2,3,4,5,6,7,8,9,10]
klus_y = [0,0,0,0,0,0,0,0,0]
sy = [0,0,0,0,0,0,0,0,0]
def find(k):
    mmu_et = 0
    ssg = 0
    for i in range(10):
        mu_p = 0
        sg = 0
        for j in range(100):
            aaa , bbb , ccc = kmean(k,a,b,100)
            ee_t              = diff(aaa,bbb,ccc)
            mu_p = mu_p + ee_t
            sg = sg + ee_t ** 2
        mu_p = mu_p / 100
        sg = sg / 100 - mu_p ** 2
        mmu_et = mmu_et + mu_p
        ssg = ssg + sg
    mmu_et = mmu_et / 10
    ssg = np.sqrt(abs(ssg / 90))
    return(mmu_et,ssg)

klus_y[0] , sy[0]  = find(2)
klus_y[1] , sy[1]  = find(3)
klus_y[2] , sy[2]  = find(4)
klus_y[3] , sy[3]  = find(5)
klus_y[4] , sy[4]  = find(6)
klus_y[5] , sy[5]  = find(7)
klus_y[6] , sy[6]  = find(8)
klus_y[7] , sy[7]  = find(9)
klus_y[8] , sy[8]  = find(10)

plt.errorbar(klus_x,klus_y,sy)
plt.show()                                # Show


```
